```python
import random
import networkx as nx

# Définir le nombre de nœuds et d'arcs
num_nodes = 20
num_arcs = 50

# Créer un graphe dirigé aléatoire avec num_nodes nœuds et num_edges arcs
Graph = nx.gnm_random_graph(num_nodes, num_arcs, directed=True)

# Ajouter des attributs aléatoires pour chaque arc
for (u, v) in Graph.edges():
    Graph.edges[u, v]['capacite'] = random.randint(1, 10)
    Graph.edges[u, v]['cout'] = random.randint(1, 10)

# Imprimer le nombre de nœuds et d'arcs du graphe
print("Nombre de nœuds : ", len(Graph.nodes()))
print("Nombre d'arcs : ", len(Graph.edges()))

# Imprimer les attributs aléatoires pour chaque arc
for (u, v) in Graph.edges():
    print("Arc ", (u, v), " : Capacité = ", Graph.edges[u, v]['capacite'], ", Coût = ", Graph.edges[u, v]['cout'])
```

```python
import matplotlib.pyplot as plt

# Extraire les capacités et les coûts des arcs
edge_labels = {(u, v): f"Capacité={Graph.edges[u, v]['capacite']}\nCoût={Graph.edges[u, v]['cout']}" for (u, v) in Graph.edges()}
capacites = [Graph.edges[u, v]['capacite'] for (u, v) in Graph.edges()]
couts = [Graph.edges[u, v]['cout'] for (u, v) in Graph.edges()]

# Dessiner le diagramme de flux
plt.figure(figsize=(40,25))
plt.title("Graphe dirigé aléatoire avec des capacités et des coûts aléatoires", fontsize=16)
plt.axis('off')
Graph = nx.DiGraph(Graph)
pos = nx.spring_layout(Graph, seed=42)
nx.draw(Graph, pos, with_labels=True, font_size=12, node_size=500)
nx.draw_networkx_edge_labels(Graph, pos, edge_labels=edge_labels, font_size=10)
for (u, v), capacite, cout in zip(Graph.edges(), capacites, couts):
    nx.draw_networkx_edge_labels(Graph, pos, edge_labels={(u,v): f"Capacité={capacite}", (v,u): f"Coût={cout}"}, font_size=10, label_pos=0.5, font_color='red', edge_color='red', edge_labels={})
plt.show()
```

```python
import random

# Définir le nombre de sources et de destinations
num_sources = 5
num_destinations = 8

# Obtenir une liste de tous les nœuds dans le graphe
nodes = list(Graph.nodes())

# Choisissez num_sources nœuds au hasard pour être des sources
Sources = random.sample(nodes, num_sources)

# Supprimer les nœuds sélectionnés de la liste des nœuds pour éviter la duplication
for node in Sources:
    nodes.remove(node)

# Choisissez num_destinations nœuds au hasard pour être des destinations
Destinations = random.sample(nodes, num_destinations)

print("Sources : ", Sources)
print("Destinations : ", Destinations)
```

```python
# Définir la valeur maximale d'offre de trafic pour les sources
max_source_offre = 50

# Affecter des valeurs d'offre de trafic aléatoires aux nœuds sources
for node in Sources:
    offre = random.uniform(0, max_source_offre)
    Graph.nodes[node]['offre'] = offre

print("Demandes d'offre de trafic des sources : ")
for node in Sources:
    print(f"Noeud {node} : {Graph.nodes[node]['offre']}")
```

```python
# Définir la valeur maximale de demande de trafic pour les destinations
max_dest_offre = 100

# Affecter des valeurs de demande de trafic aléatoires aux nœuds destinations
for node in Destinations:
    offre = random.uniform(0, max_dest_offre)
    Graph.nodes[node]['offre'] = -offre

print("Demandes de trafic des destinations : ")
for node in Destinations:
    print(f"Noeud {node} : {Graph.nodes[node]['offre']}")
```

```python
# Calculer la somme de toutes les offres et toutes les demandes
total_offre = sum(Graph.nodes[node]['offre'] for node in Sources)
total_demande = -sum(Graph.nodes[node]['offre'] for node in Destinations)

# Vérifier si la somme de l'offre est égale à la somme de la demande
if total_offre == total_demande:
    print("Le total de l'offre est égal au total de la demande.")
else:
    print("Le total de l'offre n'est pas égal au total de la demande.")
```

```python
# Créer un graphe orienté avec des attributs de capacité et de coût pour chaque arête
Graph = nx.DiGraph()

# Ajouter les arcs et leurs capacités et coûts respectifs
# Par exemple :
Graph.add_edge(1, 2, capacite=5, cout=3)

# Définir l'offre et la demande pour chaque nœud
# Par exemple :
Graph.add_node(1, demande=-3)
Graph.add_node(2, demande=3)

# Calculer le flux de coût minimum
cout_flux, dict_flux = nx.network_simplex(Graph)
```

```python
# Ajouter une contrainte de capacité pour chaque arc
for u, v, capacite in Graph.edges.data('capacite'):
    flux = dict_flux[u][v]
    if flux > capacite:
        print(f"Contrainte de capacité non respectée pour l'arc ({u}, {v})")
    else:
        print(f"Contrainte de capacité respectée pour l'arc ({u}, {v})")
```
